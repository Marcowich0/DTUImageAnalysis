```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, io, measure
from skimage.util import img_as_float
from skimage.util import img_as_ubyte

# Load the image
path = 'data/vertebra.png'
image = io.imread(path)

# Show the image
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
print(np.max(image))
print(np.min(image))

float_image = img_as_float(image)

print(np.max(float_image))
print(np.min(float_image))

ubyte_image = img_as_ubyte(float_image)
print(np.max(ubyte_image))
print(np.min(ubyte_image))
```

```python
def histogram_stretch(img_in):
    # img_as_float will divide all pixel values with 255.0
    img_float = img_as_float(img_in)
    min_val = img_float.min()
    max_val = img_float.max()
    min_desired = 0.0
    max_desired = 1.0
	
    img_out = (max_desired - min_desired) / (max_val - min_val) * (img_float - min_val) + min_desired

    # img_as_ubyte will multiply all pixel values with 255.0 before converting to unsigned byte
    return img_as_ubyte(img_out)

image_stretched = histogram_stretch(image)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot the original image
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

# Plot the stretched image
ax[1].imshow(image_stretched)
ax[1].set_title('Stretched Image')
ax[1].axis('off')

# Display the subplot
plt.show()
```

```python
def gamma_map(img, gamma):
    # img_as_float will divide all pixel values with 255.0
    img_float = img_as_float(img)
    img_out = np.power(img_float, gamma)

    # img_as_ubyte will multiply all pixel values with 255.0 before converting to unsigned byte
    return img_as_ubyte(img_out)

image_gamma = gamma_map(image, 2)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot the original image
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

# Plot the stretched image
ax[1].imshow(image_gamma)
ax[1].set_title('Gamma Image')
ax[1].axis('off')

# Display the subplot
plt.show()
```
